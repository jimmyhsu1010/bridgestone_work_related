## RT2 vs ACT in May comparison

```python
df = pd.read_clipboard()
```

```python
df["年月"] = df["年月"].astype(str)
```

```python
act = df[(df["年月"] == "202205")]
```

```python
comparison = act.copy()
```

```python
comparison.reset_index(drop=True, inplace=True)
```

```python
act.reset_index(drop=True, inplace=True)
```

```python
comparison[(comparison["GROUP別"] == "TBR") & (comparison["通路-客戶"] == "FLT-REP") & (comparison["SECTION"] == "315/80") & (comparison["花紋"] == "R294")]
```

```python
comparison[(comparison["GROUP別"] == "TBR") & (comparison["通路-客戶"] == "販社") & (comparison["花紋"].str.contains("FS|D")) & (comparison["預算/前實績"] == "22ACT")]
```

```python
comparison = comparison.drop(index=[9, 182, 252, 295, 409, 453, 460, 465, 466, 495, 524, 639, 652])
```

```python
act = act[act["GROUP別"] == "TBR"]
```

```python
comparison = comparison[comparison["GROUP別"] == "TBR"]
```

```python
result_act = act.groupby(["GROUP別", "通路-客戶", "預算/前實績"]).apply(lambda x: "{:.0f}".format(x["NET金額"].sum() / x["條數"].sum())).unstack("預算/前實績")
result_act
```

```python
act.groupby(["GROUP別", "預算/前實績"]).apply(lambda x: "{:.0f}".format(x["NET金額"].sum() / x["條數"].sum())).unstack("預算/前實績")
```

```python
result_comparison = comparison.groupby(["GROUP別", "通路-客戶", "預算/前實績"]).apply(lambda x: "{:.0f}".format(x["NET金額"].sum() / x["條數"].sum())).unstack("預算/前實績")
result_comparison
```
