```python
con1 = (sales["實際出貨日"] >= "2023-06-01")
con2 = (sales["大通路"] == "BSSC")
bssc = sales[con1 & con2]
```

```python
# 業務員為林俊良和張正平的時候，銷貨單價分別打95折和8折，只針對TBR
bssc["new_price"] = bssc.apply(lambda x: x["銷貨單價"] * 0.95 if x["業務名稱"] == "林俊良" else x["銷貨單價"] * 0.8 if x["業務名稱"] == "張正平" else x["銷貨單價"], axis=1)
```

```python
# TBR輪胎都打9折，LSR每條扣300
bssc["monthly_promo"] = bssc.apply(lambda x: x['銷貨單價'] * 0.9 if x["group"] == "TBR" else x["銷貨單價"] - 300, axis=1)
```

```python
def discount_cal(df):
    if df["group"] == "TBR" and df["業務名稱"] == "林俊良":
        return df["銷貨單價"] * 0.9 * 0.95
    elif df["group"] == "TBR" and df["業務名稱"] == "張正平":
        return df["銷貨單價"] * 0.9 * 0.8
    elif df["group"] == "TBR":
        return df["銷貨單價"] * 0.9
    else:
        return df["銷貨單價"] - 300
```

```python
bssc["discount_cal"] = bssc.apply(discount_cal, axis=1)
```

### 處理RT2數據

```python
rt2 = pd.read_clipboard()
```

```python
rt2
```

```python
filtered_rt2 = rt2[rt2["大通路"] != "BSSC"]
```

```python
filtered_rt2 = filtered_rt2.groupby(["PCODE", "新年月"])[["數量"]].sum().reset_index()
```

```python
filtered_rt2["新年月"] = filtered_rt2["新年月"].astype(str)
```

```python
filtered_rt2.columns = ["PCODE", "月份", "原數量"]
```

```python
filtered_rt2
```

```python
final = pd.merge(result, filtered_rt2, how="left")
```

```python
final = final.fillna(0)
```

```python
final["原數量"] = final["原數量"].astype(int)
```

```python
final["TS數量"] = final["數量"] - final["原數量"]
```

```python
final
```

```python
ts_vol_dict = dict(zip(zip(final["PCODE"], final["月份"]), final["TS數量"]))
```

```python
ts_vol_dict
```

```python
ts_rt2 = rt2[rt2["大通路"] == "BSSC"].drop_duplicates(subset=["新年月", "PCODE", "Type"]) # .drop_duplicates()用來丟掉重複的rows，可以加上subset參數以list的形式給出指定要檢查是否重複的欄位。
```

```python
ts_rt2["新年月"] = ts_rt2["新年月"].astype(str)
```

```python
ts_rt2["驗證用"] = ts_rt2[["PCODE", "新年月"]].apply(tuple, axis=1)
```

```python
ts_rt2
```

```python
ts_rt2["ts數量"] = ts_rt2["驗證用"].map(ts_vol_dict)
```

```python
ts_rt2 = ts_rt2[~(ts_rt2["ts數量"] <= 0)]
```

```python
ts_rt2 = ts_rt2[['新年月', 'PCODE', 'Type', '大通路', '產地', '規格', '花紋', 'ts數量', '未稅牌價', '折扣率',
       'NET', 'NET Total']].rename(columns={"ts數量": "數量"})
```

```python
other_rt2 = rt2[rt2["大通路"] != "BSSC"]
```

```python
other_rt2["新年月"] = other_rt2["新年月"].astype(str)
```

```python
final_table = pd.concat([ts_rt2, other_rt2])
```

```python
final_table["RT3 NET"] = final_table.apply(lambda x: round(x["NET"] * 0.96, 0) if x["大通路"] == "BSSC" or x["大通路"] == "AG" else x["NET"], axis=1)
```

```python
final_table["RT3折扣率"] = round((final_table["未稅牌價"] - final_table["RT3 NET"]) / final_table["未稅牌價"], 3)
```
