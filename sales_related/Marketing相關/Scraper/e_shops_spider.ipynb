### PChome價格爬蟲

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"}

init_dict = {"tires": ["https://ecshweb.pchome.com.tw/search/v3.3/all/category/DXBJ/results?q=%E8%BC%AA%E8%83%8E&page={}&sort=rnk/dc", 1], 
 "car1": ["https://ecshweb.pchome.com.tw/search/v3.3/all/category/DXBQ/results?q=%E8%BC%AA%E8%83%8E&page={}&sort=rnk/dc", 1], 
 "mi_direct": ["https://ecshweb.pchome.com.tw/search/v3.3/all/category/DXBN/results?q=%E8%BC%AA%E8%83%8E&page={}&sort=rnk/dc", 1]}
```

```python
# Get total page of each category
for i in init_dict:
    res = requests.get(init_dict[i][0].format(init_dict[i][-1]), headers=headers)
    total_page = json.loads(res.text)["totalPage"]
    init_dict[i][-1] = total_page
```

```python
init_dict
```

```python
dfs = []
for category in init_dict:
    for page in range(1, init_dict[category][-1] + 1):
        res = requests.get(init_dict[category][0].format(page), headers=headers)
        # print(init_dict[category][0].format(page))
        # print(res.status_code)
        content = json.loads(res.text)["prods"]
        df = pd.DataFrame(content)
        dfs.append(df)
        time.sleep(randint(0, 3))

result = pd.concat(dfs, ignore_index=True)
```

```python
result.to_csv(r"D:\kc.hsu\OneDrive - Bridgestone\數據\PChome_PSR價格表.csv", index=False, encoding="utf_8_sig")
```

```python
pages = get_total_page()
```

```python
def getAllPrice(pages):
    dfs = []
    today = date.today().strftime("%Y-%m-%d")
    for page in range(1, pages + 1):
        payload = {"host":"momoshop","flag":"searchEngine","data":{"specialGoodsType":"","isBrandSeriesPage":"false","authorNo":"","searchValue":"輪胎","cateCode":"3400200000","cateLevel":"1","cp":"N","NAM":"N","first":"N","freeze":"N","superstore":"N","tvshop":"N","china":"N","tomorrow":"N","stockYN":"N","prefere":"N","threeHours":"N","video":"N","cycle":"N","cod":"N","superstorePay":"N","showType":"chessboardType","curPage":"{}".format(page),"priceS":"0","priceE":"9999999","searchType":"1","reduceKeyword":"","brandName":["Michelin 米其林","Continental 馬牌","GOODYEAR 固特異","BRIDGESTONE 普利司通","DUNLOP 登祿普","PIRELLI 倍耐力","NANKANG 南港輪胎","FIREMAX 福麥斯","FALKEN 飛隼","MINERVA","NEXEN 尼克森","Altenzo"],"brandCode":["20160817181716722","20160817181652721","20160808155729906","20160817181648971","20160817181655421","20201015095812295","20210811134759748","20201119141546418","20160817181658271","20190627154027749","20220112142749731","20220112142705913"],"isFuzzy":"0","rtnCateDatainfo":{"cateCode":"3400200000","cateLv":"1","keyword":"輪胎","curPage":"1","historyDoPush":"false","timestamp":1686798093055},"flag":2018,"serviceCode":"MT01"}}
        res = requests.post(url, headers=headers, json=payload)
        content = json.loads(res.text)["rtnSearchData"]["goodsInfoList"]
        df = pd.DataFrame(content)
        dfs.append(df)
        time.sleep(randint(0, 3))
    result = pd.concat(dfs, ignore_index=True)
    result.insert(0, "UpdateDate", today)
    return result
        
```

```python
df = getAllPrice(pages)
```

```python
df["折扣方式"] = df["goodsPrice"].map(lambda x: "售價已折" if "售價" in x else "下單再折" if "下單" in x else "無")
```

```python
df["總銷量"] = df["totalSalesInfo"].map(lambda x: x["text"][3:] if len(x["text"]) != 0 else "無數據")
```

```python
df["品牌"] = df["goodsName"].map(lambda x: x.split("】")[0].replace("【", ""))
```

```python
df["幾入組"] = df["goodsName"].map(lambda x: re.findall(r"(\S)入組", x) if "車之輪" not in x else re.findall(r"(?<=-)\d", x)[-1]).map(lambda x: "".join(x))
```
