### BSSC-FLT價格整理

```python
sc_flt_price = pd.read_clipboard()
```

```python
sc_new_price = sc_flt_price[["物料", "物料.1", "交易價(未稅)"]].drop_duplicates()
```

```python
sc_new_price = sc_new_price.groupby(["物料", "物料.1"]).apply(lambda x: x["交易價(未稅)"].min()).reset_index(name="未稅最低價")
```

```python
rt3 = pd.read_clipboard()
```

```python
sc_flt_dict = dict(zip(sc_new_price["物料"], sc_new_price["未稅最低價"]))
```

```python
rt3["Current NET"] = rt3.apply(lambda x: sc_flt_dict[x["P code"]] if x["通路"] == "BSSC-FLT" else x["Current NET"], axis=1)
```
